```python
# Лабораторная работа №1
## Выполнил студент(ка) БСТ2002 Палибза Святослав Александрович
### Задание №1
import random

print("Hello, World!")


### Задание №2
def matrixGenerator(m=50, n=50, min_limit=-250, max_limit=1013):
    a = [[random.randint(min_limit, max_limit) for j in range(m)] for i in range(n)]
    print(a)
    return a

### Задание №3
def swap(mas, a, b):
    mas[a], mas[b] = mas[b], mas[a]

def swapM(mat, n, a, b):
    mat[n][a], mat[n][b] = mat[n][b], mat[n][a]


# Сортировка выбором O(m*n^2))
def selectionSort(matrix, n, m):
    for matrixString in range(m):

        for element in range(n-1):
            minIndex = element
            for i in range(element+1, n):
                if matrix[matrixString][i] < matrix[matrixString][minIndex]:
                    minIndex = i
            # ставим найденный минимум на свое место
            swapM(matrix, matrixString, minIndex, element)
    return matrix


# Сортировка вставкой O(m*n^2))
def insertionSort(matrix, n, m):
    for matrixString in range(m):
        for i in range(1, n):
            element = i
            for j in range(i):
                if matrix[matrixString][element] < matrix[matrixString][j]:
                    swapM(matrix, matrixString, element, j)
                    continue
    return matrix


# Сортировка обменом (пузырьком) O(m*n^2))
def bubbleSort(matrix, n, m):
    for matrixString in range(m):
        for i in range(n):
            iversonConditionFlag = 0
            for j in range(n - i - 1):
                if matrix[matrixString][j] > matrix[matrixString][j + 1]:
                    swapM(matrix, matrixString, j, j + 1)
                    iversonConditionFlag += 1
            if iversonConditionFlag == 0:
                return matrix
    return matrix


# Сортировка Шелла O(m*n*(log(n)^2))
def shellSort(matrix, n, m):
    for matrixString in range(m):
        d = n // 2
        while d >= 1:
            for i in range(n-d):
                if matrix[matrixString][i] > matrix[matrixString][i+d]:
                    swapM(matrix, matrixString, i, i+d)
            d //= 2

        for i in range(n):
            iversonConditionFlag = 0
            for j in range(n - i - 1):
                if matrix[matrixString][j] > matrix[matrixString][j + 1]:
                    swapM(matrix, matrixString, j, j + 1)
                    iversonConditionFlag += 1
            if iversonConditionFlag == 0:
                return matrix
    return matrix


# Турнирная сортировка O(m*n*log(n)))
def tournamentSortMatrix(matrix, n, m):
    for matrixString in range(m):
        matrix[matrixString] = tournamentSort(matrix[matrixString])
    return matrix

def tournamentSort(mass):
    final_list = []
    for i in range(len(mass)):
        final_list.append(tournamentSortSub(mass))
        mass.remove(final_list[-1])
    return final_list

def tournamentSortSub(mass):
    if len(mass) == 2:
        return min(mass)
    if len(mass) % 2 == 1:
        mass.append(2000)
    new_list = [min(mass[i], mass[i+1]) for i in range(0, len(mass)-1, 2)]
    return tournamentSortSub(new_list)



# Быстрая сортировка O(m*n*log(n)))
def quickSortMatrix(matrix, n, m):
    for matrixString in range(m):
        matrix[matrixString] = quickSort(matrix[matrixString])
    return matrix

def quickSort(mass):
    if len(mass) <= 1:
        return mass

    pivot = mass[0]
    less = [i for i in mass if mass < pivot]
    equal = [i for i in mass if i == pivot]
    more = [i for i in mass if i > pivot]

    return quickSort(less) + equal + quickSort(more)


# Пирамидальная сортировка O(m*n*log(n)))
def heapSortMatrix(matrix, n, m):
    for matrixString in range(m):
        matrix[matrixString] = heapSort(matrix[matrixString])
    return matrix


def heapSort(mass):
    # делает из массива кучу
    for j in range(len(mass) - 1, -1, -1):
        heapSortSub(mass, len(mass), j)

    for i in range(len(mass)):
        swap(mass, 0, len(mass)-i-1)
        heapSortSub(mass, len(mass) - i - 1, 0)

    return mass


def heapSortSub(mass, n, i):
    lowest = i
    left = i * 2 + 1
    right = i * 2 + 2
    if (left < n) and (mass[left] < mass[lowest]):
        lowest = left
    if (right < n) and (mass[right] < mass[lowest]):
        lowest = right
    if lowest != i:
        mass[lowest], mass[i] = mass[i], mass[lowest]
        heapSortSub(mass, n, lowest)



if __name__ == '__main__':
    try:
        m = int(input())
    except ValueError:
        m = 50
    try:
        n = int(input())
    except ValueError:
        n = 50
    try:
        min_limit = int(input())
    except ValueError:
        min_limit = -250
    try:
        max_limit = int(input())
    except ValueError:
        max_limit = 1013
    A = matrixGenerator(m, n, min_limit, max_limit)
    # print(selectionSort(A, m, n)) # сортировка выбором
    # print(insertionSort(A, 10, 3)) # сортировка вставкой
    # print(bubbleSort(A, 10, 3)) # сортировка обменом (пузырьком)
    # print(shellSort(A, 8, 1)) # сортировка Шелла
    # print(quickSortMatrix(A, 10, 3)) # быстрая сортировка
    # print(tournamentSortMatrix(A, 10, 1)) # турнирная сортировка
    print(heapSortMatrix(A, m, n)) # Пирамидальная сортировка
```
